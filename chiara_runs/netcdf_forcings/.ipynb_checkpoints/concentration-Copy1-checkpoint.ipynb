```python
import xarray as xr
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
```

```python
rcp45 = xr.open_dataset('/Users/chiaraciscato/Desktop/geomar/UVic/data.100.100.19/A_co2.nc')
ssp1 = xr.open_dataset('/Users/chiaraciscato/Downloads/A_co2_1850-2200.nc', decode_times=False)
```

```python
ssp1['A_co2'] = ssp1.A_co2 + 0 
```

```python
ssp1['time'] = ssp1.time - 0.5
```

```python
# increment_ssp1
```

```python
incr_pre = ssp1.sel(time=slice(1765.,2020.))
incr = ssp1.sel(time=slice(2021.,2200.))
```

```python
lis_incr = np.arange(.01,.46,.0025)
```

```python
incr = incr * lis_incr
```

```python
incr = ssp1.where((ssp1.time >= 2021.) & (ssp1.time <= 2200.)) + incr
```

```python
incr_fin = incr_pre.combine_first(incr)
```

```python
# decrement_rcp45
```

```python
decr_pre_01 = rcp45.sel(time=slice(1765.,2000.))
decr_01 = rcp45.sel(time=slice(2001.,2200.))
```

```python
lis_decr_01 = np.arange(0,.50,.0025)
```

```python
decr_01 = - decr_01 * lis_decr_01
```

```python
decr_01 = rcp45.where((rcp45.time >= 2001.) & (rcp45.time <= 2200.)) + decr_01
```

```python
decr_fin_01 = decr_pre_01.combine_first(decr_01)
```

```python
# decrement_rcp45
```

```python
decr_pre = rcp45.sel(time=slice(1765.,2000.))
decr = rcp45.sel(time=slice(2001.,2200.))
```

```python
lis_decr = np.arange(0,.75,.00375)
```

```python
decr = - decr * lis_decr
```

```python
decr = rcp45.where((rcp45.time >= 2001.) & (rcp45.time <= 2200.)) + decr
```

```python
decr_fin = decr_pre.combine_first(decr)
```

```python
datasets = [rcp45, ssp1, incr_fin, decr_fin_01, decr_fin]
newdata = []

for dataset in datasets:
    dataset = dataset.sel(time=slice(1850.,2200.))
    newdata.append(dataset)
```

```python
rcp45 = newdata[0]
ssp1 = newdata[1]
incr_fin = newdata[2]
decr_fin_01 = newdata[3]
decr_fin = newdata[4]
```

```python
rcp45.A_co2.plot()
ssp1.A_co2.plot()
incr_fin.A_co2.plot()
decr_fin_01.A_co2.plot()
decr_fin.A_co2.plot()
```
